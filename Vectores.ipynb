```python
import math

# Lista de vectores [x, y] de los enemigos
enemigos = [
    [3, 4], # Enemigo A
    [1, 1], # Enemigo B
    [0, 5], # Enemigo C
    [8, 6]  # Enemigo D
]

print("--- REPORTE DEL RADAR ---")

for i, pos in enumerate(enemigos):
    x = pos[0]
    y = pos[1]

    # Aplicamos Pitágoras: raíz de (x² + y²)
    distancia = math.sqrt(x**2 + y**2)

    print(f"Enemigo {i+1}: Posición {pos} -> Distancia: {distancia:.2f}")
    if distancia < 2:
        print("CUIDADO, enemigo muy cerca")
```
