Crear una variable para almacenar el nombre del usuario y otra para su edad, para mostrar en la pantalla un mensaje de bienvenida.

```python
nombre = input(" ¿Cual es tu nombre? ")
edad = int(input("¿ Cual es tu edad?"))
print(f"Hola, {nombre}! Tienes {edad} años")
```

```python
def es_mayor_de_edad(edad):
  if edad >=18:
    return "Eres mayor de edad."
  else:
    return "Eres menor de edad."
print(es_mayor_de_edad(edad))
```

Los bucles en python se utilizan para ejecutar un bloque de codigo de forma repetida.
El bucle for, se utiliza cuando sabemos donde va a terminar la repeticion. El bucle while, se utiliza cuando NO sabemos donde va a terminar

```python
for recorrer in range(5):
  print(recorrer)
```
